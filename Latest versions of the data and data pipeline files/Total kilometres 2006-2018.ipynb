```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data2 = pd.read_csv("Traffic_motor_vehicles_21102024_112422.csv", delimiter=';', decimal='.', index_col='Periods')
plt.plot(data2 [data2['Types of vehicle'].isin(['Total motor vehicles'])]
         ['Kilometres in the Netherlands/Total kilometres in the Netherlands (million km)']) #plot of the yearly million km's for all road vehicles
plt.xlabel('years')
plt.ylabel('million kilometres')
plt.title('Total kilometres travelled by motor vehicles in the Netherlands')
```
