```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Laad vier afbeeldingen
image1 = mpimg.imread('Average_kilometer_Flevoland.png')
image2 = mpimg.imread('Average_kilometer_Limburg.png')
image3 = mpimg.imread('Average_kilometer_Zeeland.png')
image4 = mpimg.imread('Average_kilometer_Noord-Brabant.png')

# Maak een figure met 2 rijen en 2 kolommen
fig, axes = plt.subplots(2, 2, figsize=(10, 10))  # 2 rijen, 2 kolommen

# Toon elke afbeelding op een aparte as
axes[0, 0].imshow(image1)
axes[0, 0].axis('off')

axes[0, 1].imshow(image2)
axes[0, 1].axis('off')

axes[1, 0].imshow(image3)
axes[1, 0].axis('off')

axes[1, 1].imshow(image4)
axes[1, 1].axis('off')

# Zorg ervoor dat er geen overlap is en ruimte goed verdeeld is
plt.tight_layout()
plt.show()
```
