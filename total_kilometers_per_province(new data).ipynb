```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("total_kilometers.csv", delimiter=';', decimal=',', skiprows=5)
#Data for the Netherlands:
tot = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Nederland'])].transpose() #total km's per year for all transport modes
tot.index = [0, 0, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017] #change index names, because names are 'mld km.1' instead of the years; first two rows are a list of names which are referring to transport mode and province

# Groningen
totGR = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Groningen (PV)'])].reset_index(drop=True).transpose()
totGR.index = tot.index

# Friesland
totFR = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Friesland (PV)'])].reset_index(drop=True).transpose()
totFR.index = tot.index

# Drenthe
totDR = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Drenthe (PV)'])].reset_index(drop=True).transpose()
totDR.index = tot.index

# Overijssel
totOV = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Overijssel (PV)'])].reset_index(drop=True).transpose() #filtering the data for Zuid-Holland
totOV.index = tot.index

# Flevoland
totFL = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Flevoland (PV)'])].reset_index(drop=True).transpose() #filtering the data for Zuid-Holland
totFL.index = tot.index

# Gelderland
totGL = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Gelderland (PV)'])].reset_index(drop=True).transpose() #filtering the data for Zuid-Holland
totGL.index = tot.index

# Utrecht
totUT = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Utrecht (PV)'])].reset_index(drop=True).transpose() #filtering the data for Zuid-Holland
totUT.index = tot.index

# Noord-Holland
totNH = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Noord-Holland (PV)'])].reset_index(drop=True).transpose() #filtering the data for Zuid-Holland
totNH.index = tot.index

# Zuid-Holland
totZH = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Zuid-Holland (PV)'])].reset_index(drop=True).transpose() #filtering the data for Zuid-Holland
totZH.index = tot.index

# Zeeland
totZE = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Zeeland (PV)'])].reset_index(drop=True).transpose() #filtering the data for Zuid-Holland
totZE.index = tot.index

# Noord-Brabant
totNB = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Noord-Brabant (PV)'])].reset_index(drop=True).transpose() #filtering the data for Zuid-Holland
totNB.index = tot.index

# Limburg
totLI = data[ data['Vervoerwijzen'].isin(['Totaal']) & data["Regio's"].isin(['Limburg (PV)'])].reset_index(drop=True).transpose() #filtering the data for Zuid-Holland
totLI.index = tot.index

# Making graph
plt.figure(figsize=(12, 8)) 
#plt.plot(totGR.iloc[2:, :], label='Groningen', marker='o')
plt.plot(totFR.iloc[2:, :], label='Friesland', marker='o')
plt.plot(totDR.iloc[2:, :], label='Drenthe', marker='o')
plt.plot(totOV.iloc[2:, :], label='Overijssel', marker='o')
plt.plot(totFL.iloc[2:, :], label='Flevoland', marker='o')
plt.plot(totGL.iloc[2:, :], label='Gelderland', marker='o')
plt.plot(totUT.iloc[2:, :], label='Utrecht', marker='o')
plt.plot(totNH.iloc[2:, :], label='Noord-Holland', marker='o')
plt.plot(totZH.iloc[2:, :], label='Zuid-Holland', marker='o')  
plt.plot(totZE.iloc[2:, :], label='Zeeland', marker='o') 
plt.plot(totNB.iloc[2:, :], label='Noord-Brabant', marker='o') 
plt.plot(totLI.iloc[2:, :], label='Limburg', marker='o') 

plt.title('Totaal mld km per jaar per provincie')
plt.xlabel('Jaar')
plt.ylabel('Totaal aantal kilometers (mld)')
plt.legend(loc='upper left')  # Legenda om onderscheid te maken tussen provincies
plt.grid(True)
plt.show()
 
```
